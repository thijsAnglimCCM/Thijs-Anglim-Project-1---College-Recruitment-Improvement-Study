1.Looking at the latest survey results (2024) from both the Majors and Non-Majors surveys, what are the gender demographic differences between the Majors and Non-Majors respondents? 

When it comes to the numerical data, more female respondents take non-majors courses than males, non-binary and not-said students combined.

```python
df24_non_genders = df24_non['gender'].value_counts()
df24_non_genders
```

```python
import matplotlib.pyplot as plt
import csv
```

Creating two lists to create variables for the categories (the keys) and their corresponding quantity (value counts) for the non-major gender column.

```python

non_gender = list(df24_non_genders.keys())
non_gender_values = list(df24_non['gender'].value_counts())
def addHlabels(x,y): #The addlabels function takes the list variables of the x and y values.
    for i in range(len(y)): #This function will iterate for every value in the y list.
        plt.text(y[i],x[i],y[i]) #For a horizontal bar graph, the first coordinate for x will be y[i]. The second coordinate for y will be x[i] meaning it will be in the position of the next x variable. The final value is the text which is the next y value.
```

```python
colors = ['pink','blue','gray','purple']
plt.barh(non_gender,non_gender_values , color=colors)
plt.title("2024 Gender Demographics for CCM Non-Major Computing Programs")
plt.xlabel("No. of students")
addHlabels(non_gender,non_gender_values)
```

According to the data, the majority of respondents are female. With 60 respondents, females are more than male, nonbinary and prefer-not-to-say respondents combined for non majors.

```python
df24_major_genders = df24_major['gender'].value_counts()
df24_major_genders
```

```python
major_gender = list(df24_major_genders.keys())
major_gender_values = list(df24_major['gender'].value_counts())

```

```python
colors = ['blue','pink','gray']
plt.barh(major_gender,major_gender_values,color = colors)
plt.title("2024 Gender Demographics for CCM Major Computing Programs")
plt.xlabel("No. of students")
addHlabels(major_gender,major_gender_values)
```

For the 2024 Computing Programming majors survey, males outnumber female and prefer-not-to-say respondents. With 162 respondents, the quanitiy of males taking a Major Computing Programming Course in CCM are more than four times the amount of respondents of other genders.

According to the demographics, most men take Majors while more women take Non-Majors than majors.
